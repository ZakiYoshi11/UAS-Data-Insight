## Visualisasi

```python
# Ubah nama tim Jerman FR menjadi Jerman
df3['Winner'] = df3['Winner'].replace('Germany FR', 'Germany')

# Hitung jumlah kemenangan setiap tim
winners_count = df3['Winner'].value_counts()

# Visualisasi
plt.figure(figsize=(10, 6))
winners_count.sort_values(ascending=False).plot(kind='bar')
plt.title('Jumlah Kemenangan Piala Dunia per Tim (1930-2014)', fontsize=16)
plt.xlabel('Tim')
plt.ylabel('Jumlah Kemenangan')
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.tight_layout()
plt.show()
```

**Noted** Visualisasi tersebut menampilkan total negara dalam menjuarai piala dunia. berikut penjelasan dari visualisasi diatas:
- Brazil berhasil menjuarai piala dunia sebanyak 5 kali
- italy dan german berhasil menjuarai piala dunia sebanyak 4 kali
- sedangkan uruguay dan argentina berhasil memenangkan piala dunia sebanyak 2 kali
- inggris, francis, dan spanyol berhasil memenangkan piala dunia sebanyak 1 kali

dari hasil visualisasi dapat diambil kesimpulan bahwa brazil dalam rentang tahun 1930-2014 menjadi team yang paling mendominasi di piala dunia, yang berarti regenarasi pemain brazil dan taktik yang diterapkan sangat efektif sehingga mampu menjuarai piala dunia sebanyak 5 kali.

```python
# Membersihkan data
df = df1.dropna()
```

```python
df['Year'] = df['Year'].astype(int)
df['Attendance'] = df['Attendance'].astype(int)
```

```python
# Mengelompokkan berdasarkan tahun dan menjumlahkan Attendance
attendance_per_year = df.groupby('Year')['Attendance'].sum().reset_index()
```

```python

# Membuat visualisasi
plt.figure(figsize=(12, 6))
plt.bar(attendance_per_year['Year'], attendance_per_year['Attendance'], 
        color='skyblue', width=0.7)

# Menambahkan judul dan label
plt.title('Total Penonton Piala Dunia per Tahun', pad=20, size=14)
plt.xlabel('Tahun', size=12)
plt.ylabel('Jumlah Penonton', size=12)

# Mengatur format sumbu x dan y
plt.xticks(attendance_per_year['Year'], rotation=45)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

# Menambahkan grid untuk memudahkan pembacaan
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mengatur layout dan margins
plt.tight_layout()

# Menampilkan grafik
plt.show()
```

**Noted**

dari hasil visualisasi menunjukkan peningkatan jumlah penonton setiap penyelenggaraan piala dunia, hal ini menunjukkan popularitas sepak bola terus meningkat. sedangkan 3 tahun pertama dalam penyelenggaraan piala dunia cendrung memiliki penonton yang rendah yaitu dibawah 1 juta penonton. setelah tahun 1938 piala dunia sempat diberhentikan dikarenakan kondisi perang dunia ke 2.penonton terbanyak piala dunia yaitu pada tahun 2014 dimana jumlah penonton lebih dari 4 juta. 

jadi dapat disimpulkan bahwa Fluktuasi dalam jumlah penonton menunjukkan bahwa ada faktor-faktor tertentu yang mempengaruhi minat penonton setiap tahunnya, seperti kualitas pertandingan, popularitas tim yang berpartisipasi, dan kondisi ekonomi global.

```python
import numpy as np

# Menyesuaikan nama tim Jerman
df3['Winner'] = df3['Winner'].replace('Germany FR', 'Germany')
df3['Runners-Up'] = df3['Runners-Up'].replace('Germany FR', 'Germany')
```

```python
# Menghitung jumlah kemenangan dan runner-up
winners_count = df3['Winner'].value_counts()
runners_up_count = df3['Runners-Up'].value_counts()
```

```python
# Mendapatkan semua tim unik
all_teams = sorted(list(set(winners_count.index) | set(runners_up_count.index)))
```

```python
# Membuat DataFrame untuk plotting
plot_data = pd.DataFrame({
    'Juara': [winners_count.get(team, 0) for team in all_teams],
    'Runner-Up': [runners_up_count.get(team, 0) for team in all_teams]
}, index=all_teams)
```

```python
# Mengurutkan berdasarkan total prestasi (juara + runner-up)
plot_data['Total'] = plot_data['Juara'] + plot_data['Runner-Up']
plot_data = plot_data.sort_values('Total', ascending=False)
plot_data = plot_data.drop('Total', axis=1)

# Membuat visualisasi
plt.figure(figsize=(12, 6))

# Posisi bar
x = np.arange(len(plot_data.index))
width = 0.35

# Membuat bars
plt.bar(x - width/2, plot_data['Juara'], width, label='Juara', color='gold')
plt.bar(x + width/2, plot_data['Runner-Up'], width, label='Runner-Up', color='silver')

# Mengatur label dan judul
plt.title('Jumlah Gelar Juara dan Runner-Up Piala Dunia per Negara (1930-2014)', 
          pad=20, size=14, fontweight='bold')
plt.xlabel('Negara', size=12)
plt.ylabel('Jumlah Gelar', size=12)

# Mengatur ticks
plt.xticks(x, plot_data.index, rotation=45, ha='right')

# Menambahkan grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Menambahkan legend
plt.legend()

# Mengatur layout
plt.tight_layout()

# Menampilkan plot
plt.show()
```

**Noted**

- Brazil: Negara dengan jumlah gelar juara terbanyak (5 gelar). Ini menunjukkan dominasi Brasil dalam sejarah Piala Dunia.

- Jerman dan Italia: Kedua negara ini juga sangat sukses di Piala Dunia dengan masing-masing 4 gelar juara. Mereka konsisten tampil kuat di turnamen ini.

- Belanda: Meskipun tidak pernah memenangkan gelar juara, Belanda memiliki jumlah gelar runner-up terbanyak (3 gelar). Ini menunjukkan bahwa mereka sering kali mencapai final tetapi belum berhasil memenangkan gelar.

- Keberagaman Pemenang: Beberapa negara seperti Uruguay, Inggris, dan Spanyol memiliki gelar juara tetapi tidak memiliki gelar runner-up, menunjukkan bahwa mereka berhasil memenangkan final ketika mereka mencapainya.

- Negara dengan Gelar Runner-Up: Negara-negara seperti Cekoslowakia, Hungaria, dan Swedia memiliki gelar runner-up tetapi tidak pernah memenangkan gelar juara, menunjukkan bahwa mereka sering kali mendekati kemenangan tetapi belum berhasil mencapainya.

```python
def label(match):
    if match["Home Team Goals"] > match["Away Team Goals"]:
        return "Tuan Rumah Menang"
    if match["Away Team Goals"] > match["Home Team Goals"]:
        return "tim Away menang"
    if match["Home Team Goals"] == match["Away Team Goals"]:
        return "seri"
```

```python
df1["Hasil Pertandingan"] = df1.apply(lambda df1:label(df1),axis=1)
plt.figure(figsize=(9,9))
df1["Hasil Pertandingan"].value_counts().plot.pie( autopct="%1.0f%%", wedgeprops={"linewidth":2})
circ = plt.Circle((0,0),.7,color="white")
plt.gca().add_artist(circ)
plt.title("Hasil pertandingan berdasarkan Home dan Away")
plt.show()
```

**Noted** 

- Tuan Rumah Menang: 57% dari pertandingan dimenangkan oleh tim tuan rumah.
- Seri: 22% dari pertandingan berakhir dengan hasil seri.
- Tim Tamu Menang: 20% dari pertandingan dimenangkan oleh tim tamu.

Ini menunjukkan bahwa tim tuan rumah memiliki keuntungan yang signifikan dalam pertandingan.
